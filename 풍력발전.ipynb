```python
from sklearn.feature_selection import mutual_info_regression
import networkx as nx
from community import community_louvain
from sklearn.metrics import mutual_info_score

target = 'Energy Production\nActive Energy Production\n[KWh]'
features = [col for col in WTG01_resampled.columns if col != target]


# 1. MI 계산 함수 (전체 데이터에 대해)
def compute_mi(WTG01_resampled, features, target):
    X = WTG01_resampled[features]
    y = WTG01_resampled[target]
    
    mi = mutual_info_regression(X, y, random_state=0, n_jobs=-1)
    return mi

# 2. MI 기반 가중치 그래프 생성
def build_mi_graph(WTG01_resampled, features):
    G = nx.Graph()
    G.add_nodes_from(features)
    # 변수쌍별 MI 계산
    for i, feat_i in enumerate(features):
        for j in range(i+1, len(features)):
            feat_j = features[j]

            xi = pd.qcut(WTG01_resampled[feat_i], q = 10, duplicates= "drop")
            xj = pd.qcut(WTG01_resampled[feat_j], q = 10, duplicates= "drop")


            mi_score = mutual_info_score(xi, xj)

        
            # MI가 0보다 크면 가중치로 추가

            if mi_score > 0:
                G.add_edge(feat_i, feat_j, weight=mi_score)

    return G




#mi_scores = compute_mi(WTG01_resampled, features, target)


#그래프 제작
G = build_mi_graph(WTG01_resampled, features)

#고립노드 제거
rm_feat = ["Availability\nForced Outage Time\n[Min.]", "Availability\nRequested Shutdown Time\n[Min.]", "Availability\nScheduled Maintenance Time\n[Min.]"]
G.remove_nodes_from(rm_feat)

partition = community_louvain.best_partition(G, weight='weight')
```

```python
plt.figure(figsize=(15, 12))
pos = nx.spring_layout(G, seed=42)  # 레이아웃 고정
cmap = plt.get_cmap('tab10')
nx.draw_networkx_nodes(G, pos, node_size=700,
                       cmap=cmap, node_color=[partition[n] for n in G.nodes()])
nx.draw_networkx_labels(G, pos)
nx.draw_networkx_edges(G, pos, edge_color='gray')
plt.title("Louvain Community Detection based on Mutual Information")
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
from collections import defaultdict
import pandas as pd

def split_dataframe_by_community(df: pd.DataFrame, partition: dict, target_col: str = None):
    # 1. community_id별 feature 정리
    community_features = defaultdict(list)
    for feat, comm_id in partition.items():
        if feat in df.columns:
            community_features[comm_id].append(feat)

    # 2. DataFrame 분리
    community_dfs = {}
    for comm_id, feats in community_features.items():
        cols = feats.copy()
        if target_col and target_col not in cols:
            cols.append(target_col)
        community_dfs[comm_id] = df[cols].copy()

    return community_dfs
```

```python
from sklearn.preprocessing import StandardScaler

feature_scaler = StandardScaler()
target_scaler = StandardScaler()

features_WTG01_0 = [col for col in WTG01_0.columns if col != 'target']
scaled_WTG01_0 = WTG01_0.copy()
scaled_WTG01_0[features_WTG01_0] = feature_scaler.fit_transform(WTG01_0[features_WTG01_0])

scaled_WTG01_0["target"] = target_scaler.fit_transform(WTG01_0[["target"]])
```

```python
import torch.nn as nn
import torch
import torch.optim as optim
import numpy as np
from torch.utils.data import Dataset, DataLoader


def create_sequences(df, sequence_length, target_col='target'):

    data = df.drop(columns=[target_col]).values
    target = df[target_col].values
    
    sequences = []
    targets = []
    
    for i in range(len(df) - sequence_length):
        seq = data[i:i+sequence_length]
        label = target[i+sequence_length]  # 다음 시점 target 예측
        sequences.append(seq)
        targets.append(label)
    
    X = np.array(sequences)
    y = np.array(targets)
    return X, y
```

```python
class TimeSeriesDataset(Dataset):
    def __init__(self, X, y):
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y, dtype=torch.float32)
    
    def __len__(self):
        return len(self.y)
    
    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]
```

```python
class LSTMRegressor(nn.Module):
    def __init__(self, input_size, hidden_size=64):
        super().__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, batch_first=True)
        self.fc = nn.Linear(hidden_size, 1)
    
    def forward(self, x):
        out, _ = self.lstm(x)
        out = out[:, -1, :]  # 마지막 타임스텝 출력
        out = self.fc(out)
        return out.squeeze()
```

```python
def train_model(df, sequence_length=6, target_col='target', epochs= 200, batch_size=32, lr=0.001):
    # 1) 시퀀스 생성
    X, y = create_sequences(df, sequence_length, target_col)
    dataset = TimeSeriesDataset(X, y)
    dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)
    
    input_size = X.shape[2]
    model = LSTMRegressor(input_size)
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=lr)
    
    model.train()
    for epoch in range(epochs):
        total_loss = 0
        for xb, yb in dataloader:
            optimizer.zero_grad()
            preds = model(xb)
            loss = criterion(preds, yb)
            loss.backward()
            optimizer.step()
            total_loss += loss.item() * xb.size(0)
        print(f"Epoch {epoch+1}/{epochs} Loss: {total_loss / len(dataset):.4f}")
    
    return model

WTG01_0_model = train_model(scaled_WTG01_0, sequence_length= 6, target_col= "target", epochs = 200, batch_size = 32, lr = 0.001)
#WTG01_1_model = train_model(WTG01_1, sequence_length= 6, target_col= "target", epochs = 200, batch_size = 32, lr = 0.001)
#WTG01_2_model = train_model(WTG01_2, sequence_length= 6, target_col= "target", epochs = 200, batch_size = 32, lr = 0.001)
```

```python

from sklearn.feature_selection import mutual_info_regression
window_size = 6
step_size = 1

mi_results = []
window_end_times = []

target = 'target'
feature = WTG01_resampled.columns.drop(target)



for start in range(0, len(WTG01_resampled) - window_size + 1, step_size):
    end = start + window_size
    window_data = WTG01_resampled.iloc[start:end]

    X = window_data[feature]
    y = window_data[target]

    mi_scores = mutual_info_regression(X, y, discrete_features = False, n_jobs = -1)
    mi_results.append(mi_scores)
    window_end_times.append(window_data.index[-1])

mi_df = pd.DataFrame(mi_results, columns = feature, index = window_end_times)

mi_df
```

```python
sliding_mi_df = mi_df.resample("D").mean()
```
