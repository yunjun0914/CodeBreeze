```python
import torch
import torch.nn as nn

def window_tensor_from_df(df, window_size = 24):
  if isinstance(df, np.ndarray):
      data_np = df
  elif isinstance(df, pd.DataFrame):
    df = df.to_numpy()
    data_np = df

  T, F = data_np.shape
  assert T % window_size == 0, f"시계열 길이 {T}는 {window_size}로 나누어 떨어져야 함"
  reshaped = data_np.reshape(-1, window_size, F)
  return torch.tensor(reshaped, dtype = torch.float32)
```

```python
import torch
import torch.nn as nn

class NodeLSTMEmbedder(nn.Module):
    def __init__(self, hidden_size=64, num_layers=1, dropout=0.0):
        super().__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.dropout = dropout
        self.lstms = None  # 나중에 initialize
        self.initialized = False

    def forward(self, x):  # x: (B, N, T, F)
        B, N, T, F_in = x.shape

        # 최초 forward에서 LSTM 구성 (lazy initialization)
        if not self.initialized:
            self.lstms = nn.ModuleList([
                nn.LSTM(F_in, self.hidden_size, self.num_layers, batch_first=True, dropout=self.dropout)
                for _ in range(N)
            ])
            self.initialized = True
        elif len(self.lstms) != N:
            raise ValueError(f"입력 노드 수 N={N}가 이전과 일치하지 않습니다. (기존: {len(self.lstms)})")

        outs = []
        for i in range(N):
            xi = x[:, i, :, :]  # (B, T, F_in)
            out, _ = self.lstms[i](xi)  # (B, T, H)
            outs.append(out.unsqueeze(1))  # (B, 1, T, H)

        data = torch.cat(outs, dim=1)  # (B, N, T, H)
        return data, data.shape[-1]
```

```python
def haversine_distance(lat1, lon1, lat2, lon2): #Haversine 거리를 이용해 인접행렬 만들기 위한 정보 생성
  R = 6371.0
  dlat = lat2 - lat1
  dlon = lon2 - lon1
  a = torch.sin(dlat / 2) **2 + torch.cos(lat1) * torch.cos(lat2) * torch.sin(dlon / 2) **2
  c = 2 * torch.atan2(torch.sqrt(a), torch.sqrt(1 - a))
  distance = R * c
  return distance
```

```python
def build_geo_adjacency(latitiudes, longitudes, threshold_km = 1.0):
  N = len(latitudes)
  lat = torch.tensor(latitudes).float() * torch.pi / 180
  lon = torch.tensor(longitudes).float() * torch.pi / 180

  lat1 = lat.unsqueeze(0).repeat(N, 1)
  lon1 = lon.unsqueeze(0).repeat(N, 1)
  lat2 = lat.unsqueeze(1).repeat(1, N)
  lon2 = lon.unsqueeze(1).repeat(1, N)

  dist_matrix = haversine_distance(lat1, lon1, lat2, lon2)
  adjacency = (dist_matrix <= threshold_km).float()
  adjacency.fill_diagonal_(1) # Self-loop 없는 버젼
  return adjacency
```

```python
class TemporalAttention(nn.Module):
  def __init__(self, hidden_size):
    super().__init__()
    self.query = nn.Linear(hidden_size, hidden_size) #[B, N, T, 1]
    self.key = nn.Linear(hidden_size, hidden_size)
    self.value = nn.Linear(hidden_size, hidden_size)


  def forward(self, x):
    B, N, T, H = x.shape
    q = self.query(x)
    k = self.key(x)
    v = self.value(x)

    q_reshaped = q.permute(0,2,1,3)
    k_reshaped = k.permute(0,2,1,3)
    v_reshaped = v.permute(0,2,1,3)

    scores = torch.matmul(q_reshaped, k_reshaped.transpose(-2, -1)) / (H ** 0.5)
    attn_weights = torch.softmax(scores, dim = 1)

    out = torch.matmul(attn_weights, v_reshaped)

    out = out.permute(0,2,1,3)
    attn_weights = attn_weights.permute(0,2,1,3)

    return out, attn_weights
```

```python
class SpatialAttention(nn.Module):
    def __init__(self, hidden_size):
        super().__init__()
        self.query = nn.Linear(hidden_size, hidden_size)
        self.key = nn.Linear(hidden_size, hidden_size)
        self.value = nn.Linear(hidden_size, hidden_size)

    def forward(self, x):
        # x: [B, N, T, H]
        B, N, T, H = x.shape
        q = self.query(x)  # [B, N, T, H]
        k = self.key(x)    # [B, N, T, H]
        v = self.value(x)

        # 내적을 위한 차원축 변형
        q_reshaped = q.permute(0, 2, 1, 3) # [B, T, N, H]
        k_reshaped = k.permute(0, 2, 1, 3) # [B, T, N, H]
        v_reshaped = v.permute(0, 2, 1, 3) # [B, T, N, H]

        # Softmax(QK^T/d^-1/2)
        scores = torch.matmul(q_reshaped, k_reshaped.transpose(-2, -1)) / (H ** 0.5) # [B, T, N, N]
        attn_weights = torch.softmax(scores, dim=2) # [B, T, N, N]

        out = torch.matmul(attn_weights, v_reshaped) # [B, T, N, H]

        out = out.permute(0, 2, 1, 3) # [B, N, T, H]
        attn_weights = attn_weights.permute(0, 2, 1, 3) # [B, N, T, H]

        return out, attn_weights
```

```python
import torch.nn.functional as F

class GCNLayer(nn.Module):
    def __init__(self, in_features, out_features):
        super().__init__()
        self.linear = nn.Linear(in_features, out_features)
        self.softplus = nn.Softplus()

    def forward(self, x, adj):
        # x: [B, N, T, H]
        # adj: [B, N, N]

        B, N, T, H = x.shape
        x = x.permute(0, 2, 1, 3)  # [B, T, N, H]

        outputs = []
        for t in range(T):
            xt = x[:, t, :, :]  # [B,T(1~24), N, H]
            # Graph convolution per time slice
            support = self.linear(xt)
            out = torch.bmm(adj, support)  # [B, N, out_features]
            outputs.append(out.unsqueeze(1))  # [B,1,N,out_features]

        out = torch.cat(outputs, dim=1)  # [B, T, N, out_features]
        out = out.permute(0, 2, 1, 3)  # [B, N, T, out_features]

        return out
```

```python
class OutputLayer(nn.Module):
    def __init__(self, input_dim, output_dim=1):
        super().__init__()
        self.fc = nn.Linear(input_dim, output_dim)
        self.relu = nn.ReLU()

    def forward(self, x):
        # x: [B, N, T, 1]
        out = self.fc(x)  # [B, N, T, 1]
        out = self.relu(out)
        return out # [B, N, T, 1]
```

```python
class FullModel(nn.Module):
    def __init__(self, latitudes, longitudes, hidden_size = 64, gcn_hidden=64):
        super().__init__()
        self.embedder = NodeLSTMEmbedder(hidden_size=hidden_size)
        self.temporal_attn = TemporalAttention(hidden_size)
        self.spatial_attn = SpatialAttention(hidden_size)
        self.gcn = GCNLayer(in_features=hidden_size, out_features=gcn_hidden)
        self.output_layer = OutputLayer(input_dim=gcn_hidden)

        lat_rad = torch.tensor(latitudes).float() * torch.pi / 180
        lon_rad = torch.tensor(longitudes).float() * torch.pi / 180
        self.register_buffer('latitudes', lat_rad)
        self.register_buffer('longitudes', lon_rad)

        A_geo = build_geo_adjacency(self.latitudes, self.longitudes).to(torch.float32)
        self.register_buffer('A_geo', A_geo)



    def forward(self, x):
        B = x.size(0)
        F_in = 43
        N = x.size(1)
        T = x.size(2)
        H = x.size(3)


        lstm_embed, _ = self.embedder(x)  # [B, N, T, H]
        temp_out, temp_weights = self.temporal_attn(lstm_embed)  # [B, N, T, H], [B, N, T, 1]
        spatial_out, spatial_weights = self.spatial_attn(lstm_embed)  # [B, N, T, H], [B, N, T, N]

        A_geo_expanded = self.A_geo.unsqueeze(0).expand(B, -1, -1)

        gcn_out = self.gcn(temp_out, A_geo_expanded)  # [B, N, T, 64]
        sum_out = torch.mean(gcn_out, dim = 1) # aggregation 방법을 mean으로 설정했으나, 더 좋은방법 필요할수도
        out = self.output_layer(sum_out)     # [B, N, T, 1]
        return out
```

```python
class EarlyStopping:
  def __init__(self, patience = 5, verbose = False, delta = 0.005):
    self.patience = patience
    self.verbose = verbose
    self.delta = delta
    self.counter = 0
    self.val_loss_min = np.inf
    self.best_score = None
    self.early_stop = False
  def __call__(self, val_loss, model):
    score = -val_loss

    if self.best_score is None:
      self.best_score = score

    elif score < self.best_score + self.delta:
      self.counter += 1
      if self.verbose:
        print(f"EarlyStopping counter: {self.counter} out of {self.patience}")
      if self.counter >= self.patience:
        self.early_stop = True
    else:
      self.best_score = score
      self.counter = 0
```
